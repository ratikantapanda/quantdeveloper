```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------
# 1 NSE Sector Indices (added Tyre)
# -------------------------
sectors = {
    "Nifty 50": "^NSEI",
    "Nifty Bank": "^NSEBANK",
    "Nifty IT": "^CNXIT",
    "Nifty Pharma": "^NSEPHARMA",
    "Nifty Auto": "^CNXAUTO",
    "Nifty FMCG": "^CNXFMCG",
    "Nifty Metal": "^NSEMETAL",
    "Nifty Energy": "^CNXENERGY",
    "Nifty Media": "^CNXMEDIA",
    "Nifty Realty": "^CNXREALTY",
    "Nifty PSU Bank": "^NSEPSUBANK",
    "Nifty Financial Services": "^NSEFIN",
    "Nifty Oil & Gas": "^CNXGAS",
    "Nifty Infrastructure": "^CNXINFRA",
    "Nifty Consumer Durables": "^CNXCD",
    "Nifty Tyre": "^CNXTYRE"
}

# -------------------------
# 2 Download Data (3 years)
# -------------------------
data = {}
for name, ticker in sectors.items():
    try:
        df_ticker = yf.download(ticker, period="3y", progress=False)
        if df_ticker.empty:
            print(f"No data for {name} ({ticker})")
            continue

        # Use Adj Close if available, else Close
        if "Adj Close" in df_ticker.columns:
            data[name] = df_ticker["Adj Close"]
        elif "Close" in df_ticker.columns:
            data[name] = df_ticker["Close"]
        else:
            print(f"No close data for {name} ({ticker})")
    except Exception as e:
        print(f"Error downloading {name} ({ticker}): {e}")

if not data:
    raise ValueError("No valid data retrieved. Check tickers or internet connection.")

# Combine into one DataFrame
df = pd.concat(data, axis=1)

# -------------------------
# 3 Calculate Daily % Change
# -------------------------
df_pct = df.pct_change().dropna() * 100

# -------------------------
# 4 Summary Table (Latest + Returns)
# -------------------------
summary = pd.DataFrame({
    "Latest Price": df.iloc[-1],
    "1M % Change": (df.iloc[-1] / df.iloc[-21] - 1) * 100 if len(df) > 21 else None,
    "3M % Change": (df.iloc[-1] / df.iloc[-63] - 1) * 100 if len(df) > 63 else None,
    "6M % Change": (df.iloc[-1] / df.iloc[-126] - 1) * 100 if len(df) > 126 else None,
    "3Y % Change": (df.iloc[-1] / df.iloc[0] - 1) * 100
})

print("=== NSE Sector Summary ===")
print(summary.round(2))

# -------------------------
# 5 Plot Normalized Sector Performance
# -------------------------
(df / df.iloc[0] * 100).plot(figsize=(14,7), linewidth=2)
plt.title("NSE Sector Indices Performance (Normalized, 3 Years)")
plt.ylabel("Index Value (Normalized to 100)")
plt.xlabel("Date")
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
